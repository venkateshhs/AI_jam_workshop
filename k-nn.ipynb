```python
clusters_to_read = [1, 2, 4, 8, 9, 10, 11, 15, 16, 17, 18, 19]
import json

train_embeddings_path = r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\example\train_embeddings.json"
test_embeddings_path = r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\example\test_embeddings.json"

with open(train_embeddings_path, "r", encoding="UTF8") as f:
    train_embeddings = json.load(f)
    
with open(test_embeddings_path, "r", encoding="UTF8") as f:
    test_embeddings = json.load(f)
base_path = r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\test\test"
document_embeddings = {}  # Dictionary to store the retrieved embeddings

for cluster in clusters_to_read:
    cluster_folder_path = os.path.join(base_path, f"cluster_{cluster}")
    
    if os.path.exists(cluster_folder_path):
        for filename in os.listdir(cluster_folder_path):
            # Extract the document ID from the filename
            doc_id, _ = os.path.splitext(filename)
            
            # Retrieve the corresponding embedding from train or test embeddings
            if doc_id in train_embeddings:
                document_embeddings[doc_id] = train_embeddings[doc_id]
            elif doc_id in test_embeddings:
                document_embeddings[doc_id] = test_embeddings[doc_id]
```

```python
import numpy as np

# Convert embeddings to a list (in the same order) for clustering
embeddings_list = list(document_embeddings.values())

# Convert the list of embeddings to a NumPy array
X = np.array(embeddings_list)
```

```python
from sklearn.cluster import KMeans

# Define the number of clusters (ensure K is defined)
K = 17  # Or any other value you're interested in

kmeans = KMeans(n_clusters=K, random_state=42, init='k-means++', algorithm="elkan").fit(X)
new_labels = kmeans.labels_
```

```python
import json
import numpy as np
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load embeddings from train and test JSON files
with open(r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\example\train_embeddings.json", "r", encoding="UTF8") as f:
    train_embeddings = json.load(f)

with open(r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\example\test_embeddings.json", "r", encoding="UTF8") as f:
    test_embeddings = json.load(f)

# Convert to numpy array
train_document_ids = list(train_embeddings.keys())
test_document_ids = list(test_embeddings.keys())
X = np.array([val for val in train_embeddings.values()] + [val for val in test_embeddings.values()])

# Perform t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Perform K-means clustering on the t-SNE reduced data
K = 17
kmeans = KMeans(n_clusters=K, random_state=60)
kmeans.fit(X_tsne)
cluster_labels = kmeans.labels_

# Plot t-SNE embeddings with different colors for different clusters
plt.figure(figsize=(10, 8))
for cluster in range(K):
    plt.scatter(X_tsne[cluster_labels == cluster, 0], X_tsne[cluster_labels == cluster, 1], alpha=0.5, label=f'Cluster {cluster}')

plt.title('t-SNE Visualization with K-means Clustering')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend()
plt.show()
plt.savefig(r"C:\Users\Vishwas\Desktop\Projects\large_language_models\AI_jam_workshop\example\results.png")


```
